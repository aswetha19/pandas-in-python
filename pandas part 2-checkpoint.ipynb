```python
import json

# Serializing the JSON.....
a={"name":"hi",
  "age":34,
  "salary":50000}
# here we can use dump() or dumps() in order to convert python object to JSON format.
b=json.dumps(a)
print(b)
print(a)
with open("data.json","w") as p:
    json.dump(a,p)
```

```python
# Deserializing JSON....
# here we can use load() and loads() in order to convert from JSON to python...
a={"name":"hi",
  "age":34,
  "salary":50000}
b=json.dumps(a)
with open("data.json","w") as p:
    json.dump(a,p)
with open("data.json","r") as q:
   c=json.load(q)
print(c)
```
